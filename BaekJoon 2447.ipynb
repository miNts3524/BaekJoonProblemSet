```python
import numpy as np

def make_rec(n,array):
    if n==3:
        array[0]=array[2]=['*','*','*']
        array[1]=['*',' ','*'] 
        return;
    
    new_n=n//3
    make_rec(new_n,array)
    for i in range(new_n):
        for j in range(new_n):
                array[i][new_n:new_n*2]=array[i][0:new_n]  
                array[i][new_n*2:new_n*3]=array[i][0:new_n] ####  (1,2) (1,3)
                
                array[i+new_n][0:new_n]=array[i][0:new_n]
                array[i+new_n][new_n*2:new_n*3]=array[i][0:new_n] #### (2,1) (2,3)
    
                array[i+new_n*2][0:new_n]=array[i][0:new_n]
                array[i+new_n*2][new_n:new_n*2]=array[i][0:new_n]
                array[i+new_n*2][new_n*2:new_n*3]=array[i][0:new_n] #### (3,1) (3,2) (3,3)

def draw(n,array):
    for i in range(n):
        for j in range(n):
            print(array[i][j],end="")
        print()
    
    
    
   






n=int(input())   # n = 3^n (n >= 1)

r_map=[[' ' for i in range (n)] for j in range (n)]

make_rec(n,r_map)
draw(n,r_map)
```
